```python
import numpy as np
import pandas as pd
import os
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
# Setup
data_root_path = "data"
impute, all_classes, primary_year = False, False, False

np.random.seed(5)
```

```python
imputed = "imputed" if impute else ""
classes = "all" if all_classes else "single"
year = "2017" if primary_year else "2016"

outfile = os.path.join(data_root_path, f"proc_data_{imputed}_{classes}_{year}.csv")
year_files = ["MERGED2016_17_PP.csv", "MERGED2017_18_PP.csv", "MERGED2018_19_PP.csv"] if primary_year else\
                ["MERGED2015_16_PP.csv", "MERGED2016_17_PP.csv", "MERGED2017_18_PP.csv"]
```

```python
df = pd.read_csv(os.path.join(data_root_path, year_files[0]))
names = [
    "UNITID",
    "INSTNM",
    "ADMCON7",
    "LOCALE",
    "RELAFFIL",
    "UGDS_WHITE",
    "CONTROL",
    "MAIN",
    "PREDDEG",
    "CCBASIC",
    "REGION",
    "HBCU",
    "ADM_RATE",
    "SATVRMID",
    "SATWRMID",
    "ACTCMMID",
    "SAT_AVG",
    "DISTANCEONLY",
    "UGDS",
    "NPT4_PUB",
    "NPT4_PRIV",
    "NUM4_PUB",
    "NUM4_PRIV",
    "COSTT4_A",
    "TUITIONFEE_IN",
    "TUITIONFEE_OUT",
    "C150_4",
    "C150_L4",
    "PCTFLOAN",
    "DEBT_MDN",
    "AGE_ENTRY",
    "FEMALE",
    "MARRIED",
    "DEPENDENT",
    "VETERAN",
    "FIRST_GEN",
    "MD_FAMINC",
    "FSEND_1",
    "MD_EARN_WNE_P10",
    "ICLEVEL",
    "OPENADMP",
    "GRADS",
    "TRANS_4_POOLED",
    "C100_4_POOLED",
    "C150_4_PELL",
    "SCHTYPE",
    "PRGMOFR",
    "CIPCODE1",
    "CIPTFBSANNUAL1",
    "MTHCMP1",
    "FTFTPCTPELL",
    "FTFTPCTFLOAN",
    "DBRR4_FED_UG_RT",
    "ROOMBOARD_ON",
    "ENDOWBEGIN",
    "MDEARN_PD",
    "COUNT_WNE_1YR",
    "CNTOVER150_1YR",
    "GT_THRESHOLD_P6",
    "MD_EARN_WNE_MALE0_P6",
    "STUFACR",
    "IRPS_WHITE",
    "IRPS_WOMEN",
    "BBRR4_FED_UG_NOPROG",
    "BBRR4_FED_UG_DISCHARGE",
    "ADM_RATE_SUPP",
    "RET_FT4"
]
features = df.loc[:, names].set_index("UNITID", drop=True)
```

```python
df2 = pd.read_csv(os.path.join(data_root_path, year_files[1]))
t_df = df2.loc[:, ["UNITID", "ADMCON7"]].set_index("UNITID", drop=True)
```

```python
df3 = pd.read_csv(os.path.join(data_root_path, year_files[2]))
o_df = df3.loc[:, ["UNITID", "RET_FT4"]].set_index("UNITID", drop=True)
```

```python
data = features.join(t_df, how="inner", rsuffix="_2").join(o_df, how="inner", rsuffix="_2")
print(data.shape)
data.head()
```

```python
# Filter for four-year not-for-profits
select = (data["CONTROL"] < 3) & (data["ICLEVEL"] == 1) & (data["DISTANCEONLY"] == 0)\
    & (data["MAIN"] == 1) & (data["OPENADMP"] == 2) & (data["REGION"] < 9)
df1 = data[select]
df1.shape
```

```python
# Remove schools which did not require testing in the baseline year
select = df1["ADMCON7"] == 1
df2 = df1[select]
df2.shape
```

```python
# Remove schools which are missing the outcome variable
select = ~df2["RET_FT4_2"].isna()
df3 = df2[select]
df3.shape
```

```python
# Remove columns which were only used to filter data
used_cols = ["MAIN", "PREDDEG", "ICLEVEL", "OPENADMP", "DISTANCEONLY", "ADMCON7"]
df4 = df3.loc[:, ~df3.columns.isin(used_cols)]
```

```python
# View missing data rates for each column
pd.set_option('display.max_rows', None)
df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).mean())
```

```python
# Identify columns completely missing data
na_pcts = df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).mean())
nan_cols = na_pcts[na_pcts == 1]
nan_cols
```

```python
# Identify columns missing more than 10% of their data
bad_cols = na_pcts[(na_pcts < 1) & (na_pcts >= 0.1)]
bad_cols
```

```python
# Select the columns with less than 10% missing data
nice_cols = na_pcts[na_pcts < 0.1].index.tolist()
nice_cols
```

```python
### NO IMPUTATION ###
# Remove schools missing data after exluding bad columns
if not impute:
    df4 = df4.loc[:, nice_cols]
    bad_schools = df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).any(), axis=1)
    df5 = df4.loc[~bad_schools, :]
    df5.shape
```

```python
### IMPUTATION ###
# Impute all remaining data after excluding bad columns
# Do not impute testing policy in treatment year
if impute:
    df4 = df4.loc[~df4["ADMCON7_2"].isna(), nice_cols]
    df4 = df4.replace(to_replace="PrivacySuppressed", value=np.nan)

    imp = IterativeImputer()
    df5 = imp.fit_transform(df4.loc[:, df4.columns != "INSTNM"])
    print(df5.shape)
    print(f"Missing data?: {np.any(np.isnan(df5))}")

    df5 = pd.DataFrame(data=np.concatenate([np.expand_dims(df4["INSTNM"], 1), df5], axis=1),
                    columns=df4.columns,
                    index=df4.index)
```

```python
### ALL CLASSES ###
# Establish treatment group as those schools which went from "required"
# to any class besides "required"
if all_classes:
    select = df5["ADMCON7_2"] != 1
    switched = df5[select]
    print(switched.shape)
    switched.head()
```

```python
### SINGLE CLASS ###
# Establish treatment group as schools which switch from "required"
# to "considered but not required". Remove schools from the other
# two classes so they are not counted in the control group
if not all_classes:
    select = df5["ADMCON7_2"] == 5
    switched = df5[select]
    df5 = df5[~((df5["ADMCON7_2"] == 2) | (df5["ADMCON7_2"] == 3))]
    print(switched.shape)
    switched.head()
```

```python
# Add treatment variable and remove baseline testing policy
df5["Z"] = np.zeros(df5.shape[0])
df5.loc[df5["INSTNM"].isin(switched["INSTNM"]), "Z"] = 1
```

```python
# Add percent change from pre to post treatment
df5 = df5[df5["RET_FT4"] != 0]
df5.loc[:, "RET_FT4_PCT_CHANGE"] = ((df5["RET_FT4_2"] - df5["RET_FT4"]) / df5["RET_FT4"]) * 100
df5 = df5.drop(columns=["RET_FT4_2"])
df5.shape
```

```python
df5.to_csv(outfile)
```

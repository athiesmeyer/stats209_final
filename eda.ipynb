```python
df2 = pd.read_csv(os.path.join(data_root_path, "MERGED2017_18_PP.csv"))
t_df = df2.loc[:, ["UNITID", "ADMCON7"]].set_index("UNITID", drop=True)
```

```python
df3 = pd.read_csv(os.path.join(data_root_path, "MERGED2018_19_PP.csv"))
o_df = df3.loc[:, ["UNITID", "RET_FT4"]].set_index("UNITID", drop=True)
```

```python
data = features.join(t_df, how="inner", rsuffix="_2").join(o_df, how="inner", rsuffix="_2")
```

```python
data.shape
```

```python
data.head()
```

```python
# Filter for four-year not-for-profits
select = (data["CONTROL"] < 3) & (data["ICLEVEL"] == 1) & (data["DISTANCEONLY"] == 0)\
    & (data["MAIN"] == 1) & (data["OPENADMP"] == 2) & (data["REGION"] < 9)
df1 = data[select]
df1.shape
```

```python
select = (~df1["ADMCON7"].isna()) & (df1["ADMCON7"] == 1)
df2 = df1[select]
print(df1.shape[0] - df2.shape[0])
df2.shape
```

```python
select = ~df2["RET_FT4_2"].isna()
df3 = df2[select]
print(df2.shape[0] - df3.shape[0])
df3.shape
```

```python
select = df3["MAIN"] == 1
df4 = df3[select]
print(df3.shape[0] - df4.shape[0])
df4.shape
```

```python
select = (df4["ADMCON7"] == 1) & (df4["ADMCON7_2"] != 1)
switched = df4[select]
print(switched.shape)
switched
```

```python
ignore_cols = ["MAIN", "PREDDEG", "ICLEVEL", "OPENADMP", "DISTANCEONLY"]
df4 = df4.loc[:, ~df4.columns.isin(ignore_cols)]
```

```python
pd.set_option('display.max_rows', None)
df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).mean())
```

```python
na_pcts = df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).mean())
nan_cols = na_pcts[na_pcts == 1]
nan_cols
```

```python
bad_cols = na_pcts[(na_pcts < 1) & (na_pcts >= 0.1)]
bad_cols
```

```python
switched.loc[:, bad_cols.index.tolist()]
```

```python
nice_cols = na_pcts[na_pcts < 0.1].index.tolist()
nice_cols
```

```python
df4 = df4.loc[:, nice_cols]
bad_schools = df4.apply(lambda x: (x.isna() | (x == "PrivacySuppressed")).any(), axis=1)
df5 = df4.loc[~bad_schools, :]
print(df5.shape)
df5.head()
```

```python
select = (df5["ADMCON7"] == 1) & (df5["ADMCON7_2"] != 1)
switched = df5[select]
print(switched.shape)
switched
```

```python
df5["Z"] = np.zeros(df5.shape[0])
df5.loc[df5["INSTNM"].isin(switched["INSTNM"]), "Z"] = 1
df5.head()
```

```python
df5.columns
```

```python
df6 = df5.loc[:, ~df5.columns.isin(["ADMCON7", "SATVRMID"])]
```

```python
df6["RET_FT4_PCT_CHANGE"] = ((df6["RET_FT4_2"] - df6["RET_FT4"]) / df6["RET_FT4"]) * 100
```

```python
df6 = df6.drop(columns=["RET_FT4_2"])
```
